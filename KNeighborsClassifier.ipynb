```python
import matplotlib.pyplot as plt
```

맷플롯립 설치

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
```

최근접 이웃 분류 알고리즘 설치

```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

도미 데이터들임

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

빙어 데이터들임

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

산점도로 도미 데이터와 빙어 데이터 표시

```python
lengths = bream_length + smelt_length
weights = bream_weight + smelt_weight
```

리스트 하나로 합치기

```python
plt.scatter(lengths, weights)
plt.xlabel("lengths")
plt.ylabel("weights")
plt.show()
```

합친거 보여주기

```python
fish_data = [[length, weight] for length, weight in zip(lengths, weights)]
fish_data[:10]
```

생선 정보를 이차원 배열으로 변환

```python
fish_target = [1] * 35 + [0] * 14
```

정답 데이터(target) 생성

1: 도미, 0: 빙어



```python
kn = KNN()
kn.fit(fish_data, fish_target) # 학습
kn.score(fish_data, fish_target) # 점수보이기
```

최근접 이웃 알고리즘으로 학습시키기

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

과연 초록색 친구를 뭐라고 예측하는지 해보자

```python
kn.predict([[30, 600]])
```

1번, 즉 도미로 올바르게 예측!

```python
kn49 = KNN(n_neighbors=49) # 참고 데이터를 49개로 설정시키기
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```
